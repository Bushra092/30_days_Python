- Create an inventory system tracking items and quantities with a dictionary


```python
inventory = {
    "milk": 20,
    "bread": 25,
    "eggs": 60,
    "rice": 100,
    "sugar": 45,
    "salt": 35,
    "butter": 15
}
```

```python

def add_item():
    item = input("Enter the item name: ").lower()
    quantity = int(input("Enter the quantity: "))
    if item in inventory :
        inventory [item] += quantity
    else:
        inventory [item] = quantity   

def remove_item():
    removable_item = input("Enter Item name You wamted to remove: ")
    if removable_item in inventory.keys():        
        print("After removed Item Your inventroy\n")
        del inventory[removable_item]  
        show_inventory()      
    else:
        print("Item not found in inventory", show_inventory())
    

def userInput():
    print("Enter 1-adding item \t 2-seeing inventory \t 3-remove items \t 4-updating \t 5-Exit")
    user_input = int(input("User input: "))
    return user_input

 
def show_inventory():
    print("Your Inventory List:",)
    for i in inventory:
        print(f"{i} \t qty:{inventory[i]}   ")


    
def main():
    print("Welcome to the Inventory Management System")
    
    while True:
        choice = userInput()
        
        if choice == 1:
            add_item()
        elif choice == 2:
            show_inventory()
        elif choice == 3:
            show_inventory()
            remove_item()
        # elif choice == 4: 
        #     changeExistItem()    
        elif choice == 5:
            print("Exiting program. Final Inventory:")
            show_inventory()
            break
        
main()


```

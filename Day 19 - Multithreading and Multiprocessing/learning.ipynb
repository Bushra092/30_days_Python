```python

#? Normal execution without Threading
import time

def cal_sqr(nums):
    print('Calculating Squares')
    for i in nums:
        time.sleep(0.2)
        print('Square:', i*i)

def cal_cub(nums):
    print('Calculating Cube')
    for i in nums:
        time.sleep(0.2)
        print('Cube:', i*i*i)        


nums_list = [2,4,6,8]

tStart = time.time()
cal_sqr(nums_list)
cal_cub(nums_list)
tEnd = time.time()
print('Done in ', tEnd - tStart)
```

```python

#? using Multi Threading
import time
import threading

square = []


def cal_sqr(nums):
    print('Calculating Squares')
    for i in nums:
        time.sleep(0.2)
        print('Square:', i*i)
        square.append(i*i)


def cal_cub(nums):
    print('Calculating Cube')
    for i in nums:
        time.sleep(0.2)
        print('Cube:', i*i*i)        


nums_list = [2,4,6,8]

tStart = time.time()

thread1 = threading.Thread(target=cal_sqr, args=(nums_list,))
thread2 = threading.Thread(target=cal_cub, args=(nums_list,))


thread1.start()
thread2.start()

thread1.join()
thread2.join()


print(square)



print('Done in ', time.time() - tStart) 
```

#### **Diffrence between them**
  explains the **difference between multiprocessing and multithreading** in computer science, especially in the context of programming and how computers handle multiple tasks.
 

### üîÑ **Multitasking**

* Doing **multiple tasks at the same time**.
  Example: A mom cooking, taking care of a baby, and answering a phone call ‚Äî all at once.
 

### üß† **Multiprocessing**

* Running **multiple processes** (separate programs) at the same time.
* Each process runs **independently** with its own **memory space**, like:

  * PowerPoint
  * Chrome
  * Calculator
  * Python script

* In Windows, you can see them in Task Manager under the ***"Details"*** tab.

üßæ Each process has its own:

* **PID (Process ID)**
* **Memory space**
* **CPU scheduling**

üîó To communicate between processes, you need **inter-process communication (IPC)** methods like:

* Files
* Shared memory
* Message pipes
 

### üßµ **Multithreading**

* A **single process** can contain **multiple threads**.
* Threads:

  * Share the **same memory space** (same process).
  * Have their **own stack**, **instruction pointer**, etc.
  * Are designed to **perform different tasks** within the same application.

üìå For example:
If you have a single process for a web browser:

* One thread might render the UI,
* Another might fetch data,
* Another might handle user inputs.

Since they share memory:

* It‚Äôs easier for them to share data.
* But, one thread's failure (like a memory leak) can crash the entire process.

 

### üîë **Main Difference**

| Feature       | Multiprocessing                         | Multithreading            |
| ------------- | --------------------------------------- | ------------------------- |
| Memory        | Separate memory per process             | Shared memory             |
| Communication | Needs IPC                               | Easier, via shared memory |
| Overhead      | High (heavyweight)                      | Low (lightweight)         |
| Isolation     | Safer (one crash doesn't affect others) | Risky (can affect others) |


### ‚úÖ Conclusion

* Use **multiprocessing** for **CPU-bound** tasks and when isolation is important.
* Use **multithreading** for **I/O-bound** or **shared-memory** tasks within the same application.

 difference between **Multithreading** and **Multiprocessing**:
